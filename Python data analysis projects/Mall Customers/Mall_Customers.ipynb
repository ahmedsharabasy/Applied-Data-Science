```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Mall_Customers.csv")

df.head()
```

```python
df.info()
```

```python
gender_v = df['Gender'].value_counts()

gender_v
```

```python
age_v = df['Spending Score (1-100)'].value_counts()

age_v
```

```python
sns.pairplot(df, hue= "Gender")
#Taking hue 
plt.show()
```

#### Good their is no outliers

```python
plt.subplots(figsize=(20, 8))
p = sns.histplot(df["Age"],kde=True,bins=50,alpha=1,fill=True,edgecolor="black")
p.axes.lines[0].set_color("#101B15")
p.axes.set_title("\nCustomer's Age Distribution\n",fontsize=25)
plt.ylabel("Count",fontsize=20)
plt.xlabel("Age",fontsize=20)
sns.despine(left=True, bottom=True)

plt.show()
```

#### The majority group in this sample is from the age of 25 to the age of 35

```python
sns.jointplot(x='Age', y='Spending Score (1-100)', data=df)

plt.show()
```
